```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Data Preprocessing

```python
df_1 = pd.read_csv('train.csv')

df_2 = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
df_1
```

```python
df_2
```

```python
df_2.rename(columns={'Attrition': 'Label'}, inplace=True)
df_2['Label'] = df_2['Label'].map({'Yes': 1, 'No': 0})
```

```python
# Calculate the count of 0s and 1s in the 'Label' column
label_counts_1 = df_1['Label'].value_counts()
```

```python
# Calculate the count of 0s and 1s in the 'Label' column
label_counts_1 = df_1['Label'].value_counts()

# Plot a pie chart
plt.figure(figsize=(6, 6))
plt.pie(label_counts_1, labels=label_counts_1.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Distribution of Labels in df_1')
plt.axis('equal')  # Ensure the pie chart is a circle
plt.show()
```

```python
# Calculate the count of 0s and 1s in the 'Label' column
label_counts_2 = df_2['Label'].value_counts()

# Plot a pie chart
plt.figure(figsize=(6, 6))
plt.pie(label_counts_2, labels=label_counts_2.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Distribution of Labels in df_2')
plt.axis('equal')  # Ensure the pie chart is a circle
plt.show()
```

```python
# Find common columns between the two dataframes
common_columns = df_1.columns.intersection(df_2.columns)

# Ensure both dataframes only contain the common columns
df_1 = df_1[common_columns]
df_2 = df_2[common_columns]

# Concatenate the dataframes
combined_data = pd.concat([df_1, df_2], ignore_index=True)
combined_data.head()
```

```python
# Display non-numeric columns
non_numeric_columns = combined_data.select_dtypes(exclude=[np.number])
print(non_numeric_columns.columns)
```

```python
# Map BusinessTravel column
travel_mapping = {'Non-Travel': 0, 'Travel_Rarely': 1, 'Travel_Frequently': 2}
combined_data['BusinessTravel'] = combined_data['BusinessTravel'].map(travel_mapping)

# Map Gender column
gender_mapping = {'Male': 0, 'Female': 1}
combined_data['Gender'] = combined_data['Gender'].map(gender_mapping)

# Map OverTime column
overtime_mapping = {'No': 0, 'Yes': 1}
combined_data['OverTime'] = combined_data['OverTime'].map(overtime_mapping)

# One-hot encode other categorical variables
combined_data = pd.get_dummies(combined_data, columns=['Department', 'EducationField', 'JobRole', 'MaritalStatus'], drop_first=True)

# Check Over18 column, and drop it if it has no information value (i.e., all values are the same)
if combined_data['Over18'].nunique() == 1:
    combined_data.drop('Over18', axis=1, inplace=True)

# Print the structure of the transformed dataframe for verification
print(combined_data.head())
```

```python
# Select boolean columns
bool_columns = combined_data.select_dtypes(include=['bool']).columns
print("Boolean columns:")
print(bool_columns)
```

```python
# Convert columns to numeric type
bool_columns = [
    'Department_Research & Development', 'Department_Sales',
    'EducationField_Life Sciences', 'EducationField_Marketing',
    'EducationField_Medical', 'EducationField_Other',
    'EducationField_Technical Degree', 'JobRole_Human Resources',
    'JobRole_Laboratory Technician', 'JobRole_Manager',
    'JobRole_Manufacturing Director', 'JobRole_Research Director',
    'JobRole_Research Scientist', 'JobRole_Sales Executive',
    'JobRole_Sales Representative', 'MaritalStatus_Married',
    'MaritalStatus_Single'
]

for col in bool_columns:
    combined_data[col] = combined_data[col].astype(bool).astype(int)
```

```python
combined_data
```

```python
combined_data.to_csv('combined_data.csv', index=False)
```

```python
# Print only the columns with missing values
missing_values = combined_data.isnull().sum()
missing_columns = missing_values[missing_values > 0]
print("Columns with Missing Values:\n", missing_columns)
```

# Correlation Analysis

```python
# Extract specified features
features = ['MonthlyIncome', 'Age', 'OverTime', 'EmployeeNumber', 'DistanceFromHome',
            'TotalWorkingYears', 'YearsAtCompany', 'PercentSalaryHike', 'JobSatisfaction',
            'NumCompaniesWorked', 'YearsWithCurrManager', 'EnvironmentSatisfaction',
            'YearsSinceLastPromotion', 'Label']

# Select the relevant columns
selected_data = combined_data[features]

# Compute the correlation matrix
correlation_matrix = selected_data.corr()

# Visualize the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Selected Features')
plt.show()
```

# Data Visualization

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Label', y='Age', data=combined_data)
plt.title('Distribution of Age by Employee Attrition')
plt.xlabel('Attrition (1=Yes, 0=No)')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Label', y='Age', data=combined_data)
plt.title('Age Distribution by Employee Attrition (Violin Plot)')
plt.xlabel('Attrition (1=Yes, 0=No)')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='OverTime', hue='Label', data=combined_data)
plt.title('OverTime Distribution by Employee Attrition')
plt.xlabel('OverTime')
plt.ylabel('Count')
plt.legend(title='Attrition (1=Yes, 0=No)')
plt.show()
```
